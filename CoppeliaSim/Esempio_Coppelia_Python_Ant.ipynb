```python
from coppeliasim_zmqremoteapi_client import RemoteAPIClient
import time
```

```python
client = RemoteAPIClient()
sim = client.getObject('sim')  # get the sim object
```

```python
sim.loadScene("C:/Users/anton/Desktop/Esempio_ant.ttt")
```

```python
Ant = sim.getObjectHandle('/Ant')
```

# Python comanda Coppeliasim tramite la funzione sim.step()

```python
sim.setStepping(True)
sim.startSimulation()

time.sleep(2)
pos = sim.getObjectPosition(Ant)
sim.setObjectPosition(Ant, [pos[0]+1, pos[1],pos[2]])
sim.step()
```
