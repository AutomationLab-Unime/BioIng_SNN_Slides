```python
from coppeliasim_zmqremoteapi_client import RemoteAPIClient
```

# Assegno handle per il client

```python
client = RemoteAPIClient()
sim = client.getObject('sim')  
```

# Carico la scena

```python
sim.loadScene("C:/Users/anton/Desktop/Esempio_cilindro.ttt")
```

```python
cilindro = sim.getObjectHandle('/Cylinder')
```

# Simulazione asincrona tra Python e Coppeliasim

```python
sim.startSimulation()
while True: 
    pos = sim.getObjectPosition(cilindro)
    sim.setObjectPosition(cilindro, [pos[0]+.0001, pos[1],pos[2]])
```
